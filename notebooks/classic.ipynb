```python
import pandas as pd
import numpy as np
```

```python
df, test_df, labels = load_dataset('./data')
df = df.drop('Unnamed: 0', axis=1)
```

```python
df
```

```python
test_df
```

```python
X = np.vstack(df['embeddings'])  # type: ignore
X_test = np.vstack(test_df['embeddings'])  # type: ignore
```

```python
from tqdm import tqdm
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(
    C=4.0, max_iter=1000, dual=False, solver='saga', penalty='l2')


for label in tqdm(labels):
    y = df[label].to_numpy().astype(bool)
    model.fit(X, y)
    test_df[label] = model.predict_proba(X_test)[:, -1]
```

```python
labels_sorted = "toxic,severe_toxic,obscene,threat,insult,identity_hate".split(
    ",")
submission = test_df[['id'] + labels_sorted]
submission.to_csv('submission.csv', index=False)
```
